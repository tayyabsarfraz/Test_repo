# Series() Function

```python
# Series() is used to create series in pandas
# series is a one dimentional adarray with axis lables
# series function return us an object.
# parameters are data , index , dtype
# index use when we want to make index any column
# dtype is used to mention the data type of the data
```

```python
list_s=[1,2,3,4,"Tayyab","Tahir"]
y=pd.Series(list_s)
print(y)
print("the type of the obtained series : ",type(y))
```

```python
# create empty series
# empty series can be created by passing the square brackets inside the series function
empty_s=pd.Series([])
print(empty_s)
```

```python
# we can change the index of the data set 
list_s=[1,2,3,4,"Tayyab","Tahir"]
y=pd.Series(list_s,index=['a','b','c','d','e','f'],name='Data values')
print(y)
```

```python
# By using dictionary we can create the series
dict_a=pd.Series({'a':1,'b':2,'c':3})
print(dict_a)
```

# Indexing and Slicing

```python
# we can get the values of series by using indexing and slicing
index_a=pd.Series([1,2,3,4,5,6,7,8,9])
print(index_a)
print()
print(index_a[0])
print(index_a[1])
print(index_a[2])
print(index_a[3])
print()
print(index_a[3:])
```

```python
# max(),min() are  used to get the maximum and minimum values
index_a=pd.Series([1,2,3,4,5,6,7,8,9])
print(max(index_a))
print()
print(min(index_a))
```

# MERGING()

```python
# MERGING FUNCTION MERG ONE OR MORE COLUMNS AND INDECIES IN DATAFRAME BY USING ONE OR MORE KEYS
```

```python
import pandas as pd
df1=pd.DataFrame({"ID":[1,2,3,4],
                 "Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"ID":[1,2,3,4],
                 "Name":["A","B","C","D"]})
df2
```

```python
pd.merge(df1,df2,on="ID")
```

```python
# by using left and right we can change the position of the columns
```

```python
pd.merge(df2,df1,on="ID",indicator=True)
```

```python
# how parameter in the merge function
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],
                 "Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"ID":[1,2,3,5],
                 "Name":["A","B","C","D"]})
df2
```

```python
# by using inner the result will be same,it will not print the common values
pd.merge(df1,df2,how="inner")
```

```python
# by using left it will print only the left missing value not the right
pd.merge(df1,df2,how="left")
```

```python
# by using right it will print only the right values and ignore the missing values
pd.merge(df1,df2,how="right")
```

```python
# outer print the values of the both
pd.merge(df1,df2,how="outer")
```

```python
# indicator indicates that from which columns the values belong
pd.merge(df1,df2,how="outer",indicator=True)
```

```python
# when mulidimensional index are present in the data then we use left_index=True and right_index=True
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],
                 "Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"ID":[5,6,7,8],
                 "Name":["A","B","C","D"]})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
# sufix parameter 
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],
                 "Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"ID":[1,2,3,4],
                 "Class":[9,10,11,12]})
df2
```

```python
pd.merge(df1,df2,on="ID",suffixes=("_Higher","_Middle"))
```

```python
# we can concat the dataframes with unequal values.
```

```python
df5=pd.DataFrame({"ID":[1,2,3,4],
                 "Name":["A","B","C","D"],
                  "Class":[5,6,7,8]})
df5
```

```python
df6=pd.DataFrame({"ID":[7,8],
                 "Name":["G","H"],
                  "Class":[11,12]})
df6
```

```python
pd.concat([df5,df6],ignore_index=True)
```

```python
# cocat unequl dataframes columns wise
pd.concat([df5,df6],axis=1)
```

```python
# join parameter 
# in join="inner" return the common values of the both dataframs.
```

```python
pd.concat([df5,df6],axis=1,ignore_index=True,join="inner")
```

```python
# in join outer return the all values of the both data frames.
pd.concat([df5,df6],ignore_index=True, axis=1,join="outer")
```

```python
# keys prameter is used to seperate the dataframes
pd.concat([df5,df6],keys=["Upper_dataframe","Lower_dataframe"])
```

```python
pd.concat([df5,df6],axis=1,keys=["First_dataframe","Second_dataframe"])
```
